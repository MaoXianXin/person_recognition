```python
import tensorflow as tf
import numpy as np
import os
```

```python
data_dir = '/home/mao/Desktop/example/jupyter_notebook/TFRecord/'
```

```python
data_path = os.listdir(data_dir)
```

```python
data_path = [os.path.join(data_dir, path) for path in data_path]
```

```python
data_path
```

```python
sess = tf.Session()
```

```python
feature = {
    'image/encoded': tf.FixedLenFeature([], tf.string),
    'image/class/label': tf.FixedLenFeature([], tf.int64),
    'image/height': tf.FixedLenFeature([], tf.int64),
    'image/width': tf.FixedLenFeature([], tf.int64),
    'image/depth': tf.FixedLenFeature([], tf.int64)
}
```

```python
filename_queue = tf.train.string_input_producer(data_path, num_epochs=1)
```

```python
filename_queue
```

```python
reader = tf.TFRecordReader()
```

```python
reader
```

```python
_, serialized_example = reader.read(filename_queue)
```

```python
serialized_example
```

```python
features = tf.parse_single_example(serialized_example, features=feature)
```

```python
features
```

```python
image = tf.decode_raw(features['image/encoded'], tf.float32)
label = tf.cast(features['image/class/label'], tf.int32)
height = tf.cast(features['image/height'], tf.int32)
width = tf.cast(features['image/width'], tf.int32)
depth = tf.cast(features['image/depth'], tf.int32)
```

```python
image, label, height, width, depth
```

```python
#image = tf.reshape(image, [1, height, width, 3])
```

```python
image
```

```python
init_op = tf.group(tf.global_variables_initializer(),
                      tf.local_variables_initializer())
```

```python
sess.run(init_op)
```

```python
coord = tf.train.Coordinator()
threads = tf.train.start_queue_runners(coord=coord, sess=sess)
```

```python
try:
    if not coord.should_stop():
        _image, _label, _height, _width, _depth = sess.run([image, label, height, width, depth])
except tf.errors.OutOfRangeError:
    print('catch OutOfRangeError')
finally:
    coord.request_stop()
    print('finish read')
coord.join(threads)
```
